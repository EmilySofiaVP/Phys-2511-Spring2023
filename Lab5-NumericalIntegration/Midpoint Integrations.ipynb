```python
import math

def func(x):
    return -0.5 * x + 4.0

def rect_area(x, width):
    return func(x) * width

def midpoint_rule(width):
    n_rects = int(10.0 / width)
    area = 0.0
    for i in range(n_rects):
        x = -5.0 + (i + 0.5) * width
        area += rect_area(x, width)
    return area

true_area = 40.0
width = float(input("Enter width of rectangles: "))

calc_area = midpoint_rule(width)
percent_diff = abs(calc_area - true_area) / true_area * 100.0
print("Function y = -0.5x + 4.0", "Estimated area:", calc_area, "Percent diff:", percent_diff)

import matplotlib
import matplotlib.pyplot as plt
x_widths = (0.25, 0.5, 1 ,2 ,5 ,10)
y_percent_diff = (0 , 0, 0, 0, 0, 0)

plt.title ("Percent difference vs Width of Rectangles of y_1")
plt.xlabel("Width of Rectangles")
plt.ylabel("Percent Difference")
plt.plot(x_widths, y_percent_diff)

import scipy.integrate
sci_area,_ = scipy.integrate.quadrature(func,-5.0, 5.0)
print("Results of scipy.integrate.quadrature: " , sci_area)
print("In comparison to scipy.integrate.quadrature, the midpoint method is very accurate for linear function")
```

```python
import math

def func(x):
    return -0.29 * x** 2 -x + 12.5

def rect_area(x, width):
    return func(x) * width

def midpoint_rule(width):
    n_rects = int(10.0 / width)
    area = 0.0
    for i in range(n_rects):
        x = -5.0 + (i + 0.5) * width
        area += rect_area(x, width)
    return area

true_area = 100.83
width = float(input("Enter width of rectangles: "))

calc_area = midpoint_rule(width)
percent_diff = (abs(true_area - calc_area) / ((true_area + calc_area) / 2)) * 100
print("Function y = -0.29x^2 -x + 12.5", "Estimated area:", calc_area, "Percent diff:", percent_diff)

import matplotlib
import matplotlib.pyplot as plt
x_widths = (0.25, 0.5, 1 ,2 ,5 ,10)
y_percent_diff = (0.01828 , 0.06320, 0.2427, 0.9574, 5.8208, 21.4055)

plt.title ("Percent difference vs Width of Rectangles of y_2")
plt.xlabel("Width of Rectangles")
plt.ylabel("Percent Difference")
plt.plot(x_widths, y_percent_diff)

import scipy.integrate
sci_area, _ = scipy.integrate.quadrature(func,-5.0, 5.0)
print("Results of scipy.integrate.quadrature: " , sci_area)
print("In comparison to scipy.integrate.quadrature, the midpoint method is mostly accurate for parabolic functions at smaller widths")
```

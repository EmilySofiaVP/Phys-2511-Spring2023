```python
import time as tm
import random 
Chute = { 18:9, 14:3, 23:17, 38:19, 32:13, 43:25, 48:26}
Ladder = { 2:21, 5:16, 7:15, 20:39, 22:35, 24:45, 27:36}

p1 = 0
p2 = 0

print("Welcome to Chutes and Ladder")
print("Let's Play :)")
print("Player 1 and 2 are at the start position\n")
player1 = input("What is your name Player 1?    ")
player2 = input("What is your name Player 2?    ")

print()

def roll():
    die = random.randint(1,6)
    return die
while (p1 < 50) and (p2 < 50):
    tm.sleep(1)
    print()
    print(player1 + "'s turn")
    tm.sleep(1)
    print("Rolling dice for " + player1 + ".......")
    tm.sleep(2)
    dice = roll()
    print(player1 + " rolled a " + str(dice))

    p1 = p1 + dice

    print(player1 + " moved to space " + str(p1))

    if p1 in Chute:
        p1 = Chute[p1]
        tm.sleep(1)
        print(player1 + " slid down to " + str(p1))

    if p1 in Ladder:
        p1 = Ladder[p1]
        tm.sleep(1)
        print(player1 + " climbed up to " + str(p1))
        
    if (p1 > 49):
        print(player1 + " REACHED THE END, " + player1 +" WINS!")
        break
    
    print()
    print()
    #player 2 turn

    tm.sleep(1)
    print(player2 + "'s turn")
    tm.sleep(1)
    print("Rolling dice for " + player2 + ".......")
    tm.sleep(2)
    dice = roll()
    print(player2 + " rolled a " + str(dice))

    p2 = p2 + dice

    print(player2 + " moved to space " + str(p2))
    

    if p2 in Chute:
        p2 = Chute[p2]
        tm.sleep(1)
        print(player2 + " slid down to " + str(p2))

    if p2 in Ladder:
        p2 = Ladder[p2]
        tm.sleep(1)
        print(player2 + " climbed up to " + str(p2))
        
 
    if (p2 > 49):
        print()
        print(player2 + " REACHED THE END, " + player2 +" WINS!")
        break
```
